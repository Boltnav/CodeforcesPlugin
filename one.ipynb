# Codeforces Data Analysis!
## Fetching details of all Contests

```python
import requests

raw_contest_list = requests.get("https://codeforces.com/api/contest.list?gym=false")
if raw_contest_list.status_code != 200:
    print("Failed to fetch contests")    

# print(raw_contest_list.content)
# print('\n')

contests = raw_contest_list.json()['result']
list_contest_details = [contest for contest in contests]
```

```python
for details in list_contest_details:
    print(details)

print(len(list_contest_details))
```

---

```python
list_contest_standing = []
for i in range(5):
    raw_contest_standings = requests.get(f"https://codeforces.com/api/contest.standings?contestId={i+1}&count=10")
    if raw_contest_standings.status_code != 200:
        print("Failed to get contest standings.")
    list_contest_standing.append(raw_contest_standings.json()['result'])
```

### List_Contest_Standing[i] properties

```python
for i in range(len(list_contest_standing)):
    print(list_contest_standing[i])
print("\n---------\n")

for key in list_contest_standing[i]['contest']:
    print(key)
print("\n---------\n")

for key in list_contest_standing[i]['problems']:
    print(key)
print("\n---------\n")

print("Participants details:")
for key in list_contest_standing[i]['rows']:
    print(key)
print("\n---------\n")
```

### Printing rating of each problem of each contest

```python
for i in range(5):
    for iter in list_contest_standing[i]['problems']:
        print(iter['index'], "-", iter['rating'])
    print('\n')
```
